# Iris dataset

```python
print(iris.DESCR)
```

```python
iris = load_iris()
iris_dataset = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_target = pd.Series(iris.target, name = 'target')
iris_dataset_with_target = pd.concat([iris_dataset, iris_target], axis = 1)
iris_dataset_with_target.head()

correlation_matrix = iris_dataset.corr()
sns.heatmap(correlation_matrix, annot = True, vmin = -1, vmax = 1, cmap = sns.color_palette("vlag", as_cmap=True)) 
plt.show()

results = pca_module.pca(iris_dataset)
print(iris_dataset.std())
```
